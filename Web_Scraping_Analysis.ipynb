```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
# URl of the wikipidea's Human body weight website (Average weight around the world By region Table)
url = 'https://en.wikipedia.org/wiki/Human_body_weight'

response = requests.get(url)
```

```python
# Importing BeautifulSoup for searching information in the HTML code
bs = BeautifulSoup(response.text, 'html.parser')
```

```python
# Extractig the table with wikitable class
tables = bs.find_all('table',{'class':'wikitable'})
```

```python
first_table = str(tables[0])
```

```python
df = pd.read_html(first_table)[0]
```

```python
df
```

```python
# Dropping the 'REF' column
df.drop(df.columns[-1],axis=1,inplace=True)
```

```python
# Checking the data type of the table
df.info()
```

```python
df.head()
```

```python
import re
```

```python

converted_value = df['Average weight'].apply(lambda x : re.findall('\d+\.\d+',x)[0]).astype('float')

df['Average weight (KG)'] = converted_value
df.head()
```

```python
converted_value_2 = df['Average weight'].apply(lambda x : re.findall('\d+\.\d+',x)[1]).astype('float')
df['Average weight (lbs)'] = converted_value_2
df.head()
```

```python
df.drop(columns='Average weight',axis=1,inplace=True)
```

```python
df.head()
```

```python
Overweight_convertion = df['% Overweight'].apply(lambda x: re.findall('\d+\.\d+',x)[0]).astype('float')
df['Overweight in %']=Overweight_convertion
```

```python
df.drop(columns='% Overweight',axis=1,inplace=True)
df.head()
```
